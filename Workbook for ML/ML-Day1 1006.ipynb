```python
import numpy as np 
from sklearn import linear_model  # scikit-learn 모듈을 가져온다

regr = linear_model.LinearRegression()

#학생들의 키 
X = [[164], [179], [162], [170]]  # 다중회귀에도 사용하도록 함 

#몸무게 
y = [53, 63, 55, 59]   # y = f(X)의 결과 

#훈련 

regr.fit(X, y)
```

```python
coef = regr.coef_ #직선의 기울기  regression 
intercept = regr.intercept_
score = regr.score(X,y)
```

```python
print("y =", coef, "* X + ", intercept)
print("The score of this line for the data: ", score)
```

```python

# 예측 대이터 
input_data = [ [180], [185] ]

#예측 
result = regr.predict(input_data)
print(result)
```

```python
# 따라하며 배우는 파이썬과 데이터과학(생능출판사 2020)
# 14.8 선형회귀로 예측하기 : 키와 몸무게는 상관관계가 있을까, 375쪽
#
import matplotlib.pyplot as plt
import numpy as np 
from sklearn import linear_model  # scikit-learn 모듈을 가져온다
 
regr = linear_model.LinearRegression() 
 
X = [[164], [179], [162], [170]]  # 선형회귀의 입력은 2차원으로 만들어야 함
#X = [164, 179, 162, 170] 

y = [53, 63, 55, 59]  

# y = f(X)의 결과값
regr.fit(X, y)

# 학습 데이터와 y 값을 산포도로 그린다. 
plt.scatter(X, y, color='black')

# 학습 데이터를 입력으로 하여 예측값을 계산한다.
y_pred = regr.predict(X)

# 학습 데이터와 예측값으로 선그래프로 그린다. 
# 계산된 기울기와 y 절편을 가지는 직선이 그려진다 
plt.plot(X, y_pred, color='blue', linewidth=3)
plt.show()
```

```python
# code_14-13

# 따라하며 배우는 파이썬과 데이터과학(생능출판사 2020)
# 14.13 알고리즘이 가지는 오차, 383쪽
#
from sklearn import datasets 
from sklearn import linear_model 
import numpy as np 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
 
# 당뇨병 데이터 세트를 sklearn의 데이터집합으로부터 읽어들인다. 
diabetes = datasets.load_diabetes() 
regr = linear_model.LinearRegression() 

X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target,
                                                    test_size = 0.2) 
regr.fit(X_train, y_train) # 학습 데이터 80% 을 학습켜서 함수발생  
y_pred = regr.predict(X_test)# 테스트 데이터 20% 

# 평균제곱오차 : MSE
print('Mean squared error:', mean_squared_error(y_test, y_pred))

```

```python
plt.scatter(y_pred, y_test,  color='black')  #x 축은 실제 데이터 , y 축은 테스트 결과 

x = np.linspace(0, 330, 100)  # 중간기준선  
plt.plot(x, x, linewidth=3, color='blue')
plt.show()
```
