```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=42) 
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
import pandas as pd 
import numpy as np 
import mglearn 
```

```python
iris_dataframe = pd.DataFrame(X_train, columns= iris_dataset.feature_names)
```

```python
print(iris_dataframe)
```

```python

grr = pd.plotting.scatter_matrix(iris_dataframe, c= y_train, figsize=(15,15), marker='o', hist_kwds={'bins':20}, s=60, alpha=.8, cmap= mglearn.cm3)

```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors=1)
```

```python
kn.fit(X_train,y_train) #kn return it's object 
```

```python
#Making prediction 

X_new = np.array([[5,2.9,1,0.2]])
print("X_new sample feature : {}".format(X_new.shape))
```

```python
prediction = kn.predict(X_new)

print("Prediction : {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_prediction = kn.predict(X_test)

print("Test set prediction:\n{}".format(y_prediction))
```

```python
print("Test set score : \n{}".format(np.mean(y_prediction == y_test)))
```

```python
print("Test set scoe: {: .2f}".format(kn.score(X_test, y_test)))
```
